2- Criar uma função que receba essa lista de 20 e-mails. A função deve percorrer a lista e, para cada e-mail, chamar a IA do Gemini via API para *resumir o conteúdo do e-mail em uma linha*, definindo o que a pessoa queria no e-mail.

```python
email_bodies = [
    "Olá, gostaria de saber se ainda há vagas disponíveis para o curso de Marketing Digital e quais são as próximas turmas. Obrigado!",
    "Boa tarde, fiz uma compra na plataforma na semana passada e ainda não recebi o código de rastreamento. Podem me ajudar?",
    "Tenho interesse em fazer uma cotação para seguro de vida familiar. Podem enviar os valores e planos disponíveis?",
    "Olá, tentei acessar minha conta mas aparece que meu e-mail não está cadastrado. Podem verificar o que aconteceu?",
    "Estou precisando solicitar a segunda via da minha fatura. O app não está permitindo gerar. Conseguem emitir manualmente para mim?",
    "Gostaria de cancelar minha assinatura. Não estou utilizando o serviço e desejo encerrar o plano ainda este mês.",
    "Vocês oferecem suporte para integrar o sistema de vocês com o CRM HubSpot? Minha empresa quer conectar as duas plataformas.",
    "Tenho interesse em saber quando será lançado o novo modelo do smartphone X e qual será o preço estimado.",
    "Fiz a devolução do produto conforme orientado, mas ainda não recebi o estorno no cartão. Quanto tempo leva?",
    "Gostei muito do atendimento na loja ontem e gostaria de deixar registrado um elogio para o vendedor João.",
    "Gostaria de solicitar um orçamento para criação de um site institucional com até 5 páginas e blog integrado.",
    "Meu cachorro está com vômito e diarreia desde ontem. Gostaria de saber se há veterinário disponível hoje à tarde.",
    "Preciso alterar o endereço de entrega do pedido #28374. Ele foi feito ontem e ainda não saiu para envio.",
    "Quando será liberado o resultado do processo seletivo? Participei da entrevista há duas semanas e ainda não recebi retorno.",
    "Estou com dificuldades para configurar a internet fibra. O modem mostra luz vermelha no indicador. Podem avaliar remotamente?",
    "Estou planejando viajar para o Chile em maio. Gostaria de uma cotação de passagem aérea e hospedagem.",
    "O relatório financeiro mensal veio com resultados incoerentes. Alguma atualização pode ter afetado os dados?",
    "Parabéns pelo evento! Foi uma experiência incrível. Existe como receber os slides das palestras apresentadas?",
    "Gostaria de incluir um dependente no meu plano de saúde. Como funciona o processo e quais documentos preciso enviar?",
    "Realizei transferência via PIX para a conta da empresa ontem à noite, mas o pagamento ainda consta como pendente na plataforma."
]
```

```python
def resumir_emails(lista_emails):
  for numero, email in enumerate(lista_emails):
    resposta = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=f'Vou te enviar 20 e-mails.Resuma o que a pessoa quer em cada e-mail, em exatamente 1 linha por e-mail.{email}'
        )
    print(f' Email {numero}: {resposta.text}')
    print('-' * 50)
```

```python
resumir_emails(email_bodies)
```
